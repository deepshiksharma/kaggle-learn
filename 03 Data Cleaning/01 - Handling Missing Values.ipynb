# [Handling Missing Values](https://www.kaggle.com/code/alexisbcook/handling-missing-values)

```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)
```

```python
# Data source: https://www.kaggle.com/code/rtatman/data-cleaning-challenge-handling-missing-values/input?select=NFL+Play+by+Play+2009-2017+%28v4%29.csv
nfl_data = pd.read_csv('NFL Play by Play 2009-2017 (v4).csv')
```

```python
nfl_data.head()
```

## Checking for missing data points

```python
# Number of missing data points per column
missing_values_count = nfl_data.isnull().sum()

# Number of missing points in the first ten columns
missing_values_count[0:10]
```

```python
# Total number of missing values
total_cells = np.product(nfl_data.shape)
total_missing = missing_values_count.sum()

# Percentage of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

## Drop missing values

`panda`'s built-in `dropna()` function drops rows or columns that have missing values.

```python
# Remove all rows with at least one missing value
nfl_data.dropna()
```

The operation above removed all data points because every row in the dataset had atleast one missing value. <br>
We might have better luck by instead removing all columns that have at least one missing value.

```python
# Remove all columns with at least one missing value
nfl_data.dropna(axis=1)
```

```python
columns_with_na_dropped = nfl_data.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
# How much data did we lose?
print('Columns in original dataset:', nfl_data.shape[1])
print('Columns with na\'s dropped:', columns_with_na_dropped.shape[1])
```

We've lost quite a bit of data, but at this point we have successfully removed all the `NaN`'s from our data.

## Filling in missing values automatically

```python
# A small subset of the NFL dataset
subset_nfl_data = nfl_data.loc[:, 'EPA':'Season'].head()
subset_nfl_data
```

`panda`'s built-in `fillna()` function fills in missing values in a DataFrame. <br>
One option we have is to specify what we want the NaN values to be replaced with. Here all NaN values are replaced with 0.

```python
# Replace all NA's with 0
subset_nfl_data.fillna(0)
```

Backfill `bfill` and Forward fill `ffill` can also be used to fill in missing values.

```python
# Replace all NA's with the value that comes directly after it in the same column, then replace all the remaining na's with 0
subset_nfl_data.fillna(method='bfill', axis=0).fillna(0)
```

```python
# Replace all NA's with the value that comes directly before it in the same column, then replace all the remaining na's with 0
subset_nfl_data.fillna(method='ffill', axis=0).fillna(0)
```

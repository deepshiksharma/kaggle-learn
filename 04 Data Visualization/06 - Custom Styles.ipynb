```python
import pandas as pd
import matplotlib.pyplot as plt

pd.plotting.register_matplotlib_converters()
%matplotlib inline

import seaborn as sns
```

```python
spotify_data = pd.read_csv('data/spotify.csv', index_col='Date', parse_dates=True)
spotify_data.head()
```

```python
sns.lineplot(data=spotify_data)
```

## Changing plot styles
Seaborn has five different themes: <br> 
(1)`"darkgrid"`, (2)`"whitegrid"`, (3)`"dark"`, (4)`"white"`, and (5)`"ticks"`. <br> 
You need only use a command similar to the one in the code cell below to change it.

```python
# Change the style of the figure to a "dark" theme
sns.set_style('dark')
sns.lineplot(data=spotify_data)
```

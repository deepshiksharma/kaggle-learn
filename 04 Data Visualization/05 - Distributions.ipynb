```python
import pandas as pd
import matplotlib.pyplot as plt

pd.plotting.register_matplotlib_converters()
%matplotlib inline

import seaborn as sns
```

We'll work with a dataset of 150 different flowers; 50 each from three different species of iris (Iris setosa, Iris versicolor, and Iris virginica).

```python
iris_data = pd.read_csv('data/iris.csv', index_col='Id')
iris_data.head()
```

## Histogram

```python
sns.histplot(iris_data['Petal Length (cm)'])
```

## Density plot
This type of plot is a kernel density estimate (KDE) plot. You can think of a KDE plot as a smoothed histogram. <br>

To make a KDE plot, use the `sns.kdeplot` command. <br>
Setting `fill=True` colors the area below the curve, and `data=` chooses the column we would like to plot.

```python
sns.kdeplot(data=iris_data['Petal Length (cm)'], fill=True)
```

## 2D KDE plot
We're not restricted to a single column when creating a KDE plot. We can create a two-dimensional KDE plot with the `sns.jointplot` command. <br>

In the plot below, the color-coding shows us how likely we are to see different combinations of sepal width and petal length, where darker parts of the figure are more likely.

```python
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind='kde')
```

Note that in addition to the 2D KDE plot in the center,
<ul>
    <li>the curve at the top of the figure is a KDE plot for the data on the x-axis. iris_data['Petal Length (cm)'] in this case. </li>  
    <li>the curve on the right of the figure is a KDE plot for the data on the y-axis. iris_data['Sepal Width (cm)'] in this case. </li>
</ul>

## Color-coded plots
We'll create plots to understand differences between the species. <br>
We can create three different histograms (one for each species) of petal length by using the `sns.histplot` command. <br>

`data=` provides the name of the variable that we used to read in the data <br>
`x=` sets the name of column with the data we want to plot <br>
`hue=` sets the column we'll use to split the data into different histograms

```python
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species')
plt.title('histogram of petal lengths, by species')
```

We can also create a KDE plot for each species by using `sns.kdeplot` command. <br> 

The functionality for `data`, `x`, and `hue` are identical to `sns.histplot` above. Additionally, we set `fill=True` to color the area below each curve.

```python
sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', fill=True)
plt.title('distribution of petal lengths, by species')
```

An interesting pattern that can be seen in plots is that the plants seem to belong to one of two groups, where `Iris versicolor` and `Iris virginica` seem to have similar values for petal length, while `Iris setosa` belongs in a category all by itself. <br>

According to this dataset, we might even be able to classify any iris plant as Iris setosa (as opposed to Iris versicolor or Iris virginica) just by looking at the petal length: if the petal length of an iris flower is less than 2 cm, it's most likely to be Iris setosa!

## Selecting

A specific Series from a DataFrame can be accessed by indexing into the column name, like this:

```python
reviews['country']
```

A DataFrame can be thought of as an object, with its columns as properties. Hence, we can access the property of an object by accessing it as an attribute. <br>
The `country` property of `reviews` can be accessed like this:

```python
reviews.country
```

To access one specific value from a Series in a DataFrame, we can index into it a step further like this:

```python
reviews['country'][1]
```

## Indexing

pandas has its own accessor operators, `iloc` and `loc`. For advanced operations, these are the ones you're supposed to be using. <br>
pandas indexing works in one of two paradigms, <b>index-based selection</b> and <b>label-based selection</b>.

### Index-based selection
Index-based selection: selecting data based on its numerical position in the data. `iloc` follows this paradigm.

To retrieve the first row of a DataFrame:

```python
reviews.iloc[0]
```

To retrieve the first column of a DataFrame:

```python
reviews.iloc[:, 0]
```

To retrieve the country column from just the first, second, and third rows:

```python
reviews.iloc[:3, 0]
```

Or, to To retrieve the country column from just the second and third rows:

```python
reviews.iloc[1:3, 0]
```

It is also possible to pass in a list of indices to retrieve the desired rows and/or columns:

```python
reviews.iloc[[0, 1, 2, 3, 4], [0, 2]]
```

Just like with Python list indexing, negative indices can be used to retrieve data from the end:

```python
reviews.iloc[-5:, 0]
```

### Label-based selection
Label-based selection: selecting data based on its index value, not position. `loc` follows this paradigm.

To retrieve the `country` from the first entry in `reviews`:

```python
reviews.loc[0, 'country']
```

```python
reviews.loc[1, ['country', 'designation', 'province', 'taster_name', 'taster_twitter_handle', 'title', 'variety', 'winery']]
```

### An important note on the difference between `iloc` and `loc` indexing schemes

`iloc` uses the Python stdlib indexing scheme, where the first element of the range is included and the last one is excluded. <br>
This means `iloc` `0:10` will select entries 0,...,9. <br>

Whereas `loc` indexes inclusively. This means `loc` `0:10` will select entries 0,...,10.

```python
reviews.iloc[0:10, 0]
```

```python
reviews.loc[0:10, 'country']
```

## Conditional selection

Suppose that we're interested specifically in better-than-average wines produced in Italy.

We can start by checking if each wine is Italian or not:

```python
reviews.country == 'Italy'
```

The above operation produced a Series of True/False booleans based on the country of each record. <br>
This result can then be used inside of loc to select the relevant data:

```python
reviews.loc[reviews.country == 'Italy']
```

We also wanted to know which ones are better than average. <br>
Wines are reviewed on a 80-to-100 point scale, so this could mean wines that accrued at least 90 points. <br>

We can use the ampersand (&) to bring the two questions together:

```python
reviews.loc[(reviews.country == 'Italy') & (reviews.points >= 90)]
```

Suppose we'll buy any wine that's made in Italy or which is rated above average. For this we use the pipe (`|`) operator:

```python
reviews.loc[(reviews.country == 'Italy') | (reviews.points >= 90)]
```

### Built-in conditional selectors in pandas
pandas comes with a few built-in conditional selectors, two of which will be highlighted here.

`isin` lets you select data whose value "is in" a list of values. <br>
Here's how we can use it to select wines only from Italy or France:

```python
reviews.loc[reviews.country.isin(['Italy', 'France'])]
```

`isnull` and its companion `notnull` lets you highlight values which are (or are not) empty (NaN). <br>
To filter out wines lacking a price tag in the dataset, here's what we would do:

```python
reviews.loc[reviews.price.notnull()]
```

Alternatively, `isnull` can be used to display all the wines lacking a price tag.

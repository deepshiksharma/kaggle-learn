```python
import pandas as pd
```

```python
wine_data = pd.read_csv('data/winemag-data-130k-v2.csv', index_col=0)
print(wine_data.shape)

wine_data.head()
```

### Create a new data frame containing geographic information after conditional selection based on wine points

```python
geo_data = wine_data[wine_data['points'] > 90][['country', 'province']].copy()
print(geo_data.shape)

geo_data.head()
```

```python
geo_data = wine_data.loc[
    wine_data['points'] > 90, ['country', 'province']
].copy()

geo_data.tail()
```

### Create a new data frame containing taster's name and twitter handle if province == 'California'

```python
# '\' is the line continuation character in Python
# To improve readability, the code can be written across multiple lines.

taster_data = wine_data[
    wine_data['province'] == 'California'
] \
[['taster_name', 'taster_twitter_handle']].copy()

print(taster_data.shape)
```

```python
taster_data.head()
```

```python
taster_data = wine_data.loc[
    wine_data['province'] == 'California', ['taster_name', 'taster_twitter_handle']
].copy()

taster_data.tail()
```
